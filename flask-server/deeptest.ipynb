```python
from DeepImageSearch import Index,LoadData,SearchImage
```

```python

img_list = LoadData().from_folder(['house_img'])
img_list
```

```python
Index(img_list).Start()
```

```python
ordered_imgs = SearchImage().get_similar_images(image_path='house_img/house_test_2.jpg', number_of_images=5)
ordered_imgs
```

```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg


l = len(ordered_imgs)
f, axarr = plt.subplots(l, l)

r, c = 0, 0
for i in ordered_imgs:
    img = mpimg.imread(img_list[i])
    axarr[r,c].imshow(img)
    c += 1
    if c == l:
        c = 0
        r += 1
```
